```python
import pandas as pd
```

```python
dfAcre = pd.read_csv('Acre.csv', sep = ';')
dfAlagoas = pd.read_csv('Alagoas.csv', sep = ';')
dfAmazonas = pd.read_csv('Amazonas.csv', sep = ';')
dfBahia = pd.read_csv('Bahia.csv', sep = ';')
dfCeara = pd.read_csv('Amazonas.csv', sep = ';')
dfDF = pd.read_csv('DF.csv', sep = ';')
dfEspiritoSanto = pd.read_csv('EspiritoSanto.csv', sep = ';')
dfGoias = pd.read_csv('Goias.csv', sep = ';')
dfMaranhao = pd.read_csv('Maranhao.csv', sep = ';')
dfMatoGrosso = pd.read_csv('MatoGrosso.csv', sep = ';')
dfMatoGrossoDoSul = pd.read_csv('MatoGrossoDoSul.csv', sep = ';')
dfMinasGerais = pd.read_csv('MinasGerais.csv', sep = ';')
dfPara = pd.read_csv('Para.csv', sep = ';')
dfParana = pd.read_csv('Parana.csv', sep = ';')
dfPernambuco = pd.read_csv('Pernambuco.csv', sep = ';')
dfPiaui = pd.read_csv('Piaui.csv', sep = ';')
dfRioDeJaneiro = pd.read_csv('RioDeJaneiro.csv', sep = ';')
dfRioGrandeDoNorte = pd.read_csv('RioGrandeDoNorte.csv', sep = ';')
dfRioGrandeDoSul = pd.read_csv('RioGrandeDoSul.csv', sep = ';')
dfRondonia = pd.read_csv('Rondonia.csv', sep = ';')
dfRoraima = pd.read_csv('Roraima.csv', sep = ';')
dfSantaCatarina = pd.read_csv('SantaCatarina.csv', sep = ';')
dfSergipe = pd.read_csv('Sergipe.csv', sep = ';')
dfSPCapital = pd.read_csv('SPCapital.csv', sep = ';')
dfSPExcetoCapital = pd.read_csv('SPExcetoCapital.csv', sep = ';')
dfTocantins = pd.read_csv('Tocantins.csv', sep = ';')



```

```python
dfTotal = pd.concat([dfAcre, dfAlagoas, dfAmazonas, dfBahia, dfCeara, dfDF, dfEspiritoSanto, dfGoias, dfMaranhao, dfMatoGrosso, dfMatoGrossoDoSul,
           dfMinasGerais, dfPara, dfParana, dfPernambuco, dfPiaui, dfRioDeJaneiro, dfRioGrandeDoNorte, dfRioGrandeDoSul, dfRondonia, dfRoraima,
           dfSantaCatarina, dfSergipe, dfSPCapital, dfSPExcetoCapital, dfTocantins
           ], ignore_index=True)
```
